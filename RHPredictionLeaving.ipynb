## Visualização dos dados

Convertendo dados binários que estão como string para 0/1

```python
for column in ["Attrition", "OverTime"]:
    employee_df[column] = employee_df[column].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x == 'Y' else 0)
```

```python
employee_df.head()
```

Separando dados dos funcionários que deixaram e os que continuaram na empresa

## Pré-processamento e criação das bases de treinamento e teste

```python
employee_df.head()
```

Features categóricas que devem ser transformadas

```python
X_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
X_cat
```

Utilizando o OneHotEncoder para transformar as features categoricas

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
employee_df['BusinessTravel'].unique()
```

```python
X_cat
```

Features númericas que não foram trasnformadas pois não precisam

```python
X_numerical = employee_df[['Age', 'DailyRate', 'DistanceFromHome',	'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement',	'JobLevel',	'JobSatisfaction',	'MonthlyIncome',	'MonthlyRate',	'NumCompaniesWorked',	'OverTime',	'PercentSalaryHike', 'PerformanceRating',	'RelationshipSatisfaction',	'StockOptionLevel',	'TotalWorkingYears'	,'TrainingTimesLastYear'	, 'WorkLifeBalance',	'YearsAtCompany'	,'YearsInCurrentRole', 'YearsSinceLastPromotion',	'YearsWithCurrManager']]
X_numerical
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
X_all.columns
```

Separação do dataset em dados de treinamento e teste

```python
from sklearn.model_selection import train_test_split
y = employee_df['Attrition']
X_train, X_test, y_train, y_test = train_test_split(X_all, y, test_size = 0.10, random_state=42)
```

Normalizando os dataset, utilizando o dataset de treinamento para que não fique implicito para o algoritmo os máximos e mínimos do conjunto de teste

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape, y_train
```

```python
X_test.shape, y_test
```

## Regressão logística

Utilizando o algoritmo de regressão logistica para tentar fazer a previsão

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
y_pred
```

```python
y_test
```

Calculando a acurácia para o problemas

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

Calculando a matriz de confusão

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True);
plt.show()
```

Calculando outras métricas

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred, average='macro')
```

- Precision: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_score.html
- Recall: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.recall_score.html
- F1: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html

```python
print(classification_report(y_test, y_pred))
```

## Random forest

Utilizando o algoritmo Random Forest para resolver o problema

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
y_pred = forest.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_pred, y_test)
cm
```

```python
sns.heatmap(cm, annot=True);
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

## Redes neurais artificiais

Utilizando uma rede neural do tipo MLP para resolver o problema

```python
import tensorflow as tf
```

```python
rede_neural = tf.keras.models.Sequential()
rede_neural.add(tf.keras.layers.Dense(units = 64, activation='relu', input_shape=(50,)))
rede_neural.add(tf.keras.layers.Dense(units = 32, activation = 'relu'))
rede_neural.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
rede_neural.summary()
```

```python
rede_neural.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
rede_neural.fit(X_train, y_train, epochs=50, verbose=2)
```

```python
y_pred = rede_neural.predict(X_test)
y_pred = (y_pred >= 0.5)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True);
plt.show()
```

Todas as métricas da rede para o problema

```python
print(classification_report(y_test, y_pred))
```
